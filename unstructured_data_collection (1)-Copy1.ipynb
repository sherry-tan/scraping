```python
import requests
from bs4 import BeautifulSoup


def get_soup(url):

    header = {
        "User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.108 Safari/537.36Name"
    }

    content = requests.get(url , headers = header ).text

    soup = BeautifulSoup(content, "html.parser")
    
    return soup

searchlisting_url = "https://en.jobs.sanofi.com/search-jobs/Singapore?orgIds=20873&alp=1880251&alt=2"

listingsoup = get_soup(searchlisting_url)

```

```python
jobpostings= listingsoup.find("ul", {"class": "unstyled-list job-list"})

url_prefix = "https://en.jobs.sanofi.com/"
url_list=[]
for posting in jobpostings.find_all('a', href=True):
    url = url_prefix+posting['href']
    url_list.append(url)
    
url_list
```

```python
jobinfo=[]

for url in url_list:
    soup = get_soup(url)
    postInfo = {}
    postInfo["role"] = soup.find("title").text
    postInfo["datePosted"] = soup.find('span',{'class':"job-date job-info icon-time with-text"}).text
    postInfo["contractType"] = soup.find('span',{'class':"icon-contract with-text"}).text
    postInfo["jobID"] = soup.find("span", {'class':"job-id job-info icon-job-id with-text"}).text
    postInfo["jobLocation"] = soup.find("span",{'class':"job-location job-info white-text"}).text.strip()
    desc = soup.find('div',{'class':"ats-description"})
    
    sections=desc.find_all('p')

    for section in sections:
        if section.find('ul'):
            #sectionName = section.find('p').text
            sectionName = section.find('b').text
            postInfo[sectionName] = [listitem.text.strip() for listitem in section.find_all('li')]

    jobinfo.append(postInfo)
```
